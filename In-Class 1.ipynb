```python
from bs4 import BeautifulSoup

import urllib.request

import nltk

from nltk.corpus import stopwords

#Opening the spaceX wikipedia's website
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')

#Reading the contents of the site and storig it in a variable named "html"
html = response.read()

#Creating an object for the BeautifulSoup class oriented towards html5
soup = BeautifulSoup(html,"html5lib")

#Eliminating any html5 elements like photos, advertisements etc. and just retrieving text
text = soup.get_text(strip=True)

#Separating the words in sentences
tokens = [t for t in text.split()]

#Clean_tokens in order to utilize after removing stop words
clean_tokens = tokens[:]

#Retrieving the list of stopwords
sr = stopwords.words('english')

#Removing stopwords from all the tokens and storing it in clean_tokens
for token in tokens:

    if token.lower() in sr:

        clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)
```

```python
import matplotlib.pyplot as plt

# newfreq dictionary to separate the words with higher than or equal to 5 frequency/occurrence
newfreq = {}
for key,val in freq.items():
    if val >= 5:
        newfreq[key] = val

# Dictionary to store the key-value pairs of top-10 high frequency words        
finaldict = {}

# The top 10 high-frequency values
top10 = sorted(newfreq.values(), reverse = True)[:10]

# Sorting and creating a new dictionary for the top 10 high-frequency words
for i in top10:
    for key, val in newfreq.items():
        if newfreq[key] == i:
            finaldict[key] = val
            break

#Line plot to visually represent the top 10 high-frequency words

plt.xticks(rotation = 90)
plt.xlabel("\nWords")
plt.ylabel("Frequency")
plt.title("Top 10 words in the SpaceX Wikipedia webpage based on their frequency\n\n")
plt.plot(finaldict.keys(), finaldict.values())
```

```python
#Bar plot to visually represent the top 10 high-frequency words

plt.xticks(rotation = 90)
plt.xlabel("\nWords")
plt.ylabel("Frequency")
plt.title("Top 10 words in the SpaceX Wikipedia webpage based on their frequency\n\n")
plt.bar(finaldict.keys(), finaldict.values())
```
